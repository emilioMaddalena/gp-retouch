```python
# Visualizing some kernel functions
```

```python
import GPy

from gp_retouch.image.image_processor import ImageProcessor
from gp_retouch.artist.artist import Artist
```

```python
kernel = GPy.kern.RBF(input_dim=2, variance=200.0, lengthscale=10.0)
image = Artist.draw_kernels(kernel, num_centers=10).plot()
```

```python
kernel = GPy.kern.Linear(input_dim=2)
image = Artist.draw_kernels(kernel, num_centers=5).plot()
```

```python
kernel = (
    GPy.kern.PeriodicExponential(input_dim=1, active_dims=0, lengthscale=10.0, period=10) 
    + GPy.kern.PeriodicExponential(input_dim=1, active_dims=1, lengthscale=2.0, period=10)
)
Artist.draw_kernels(kernel, num_centers=4).plot()
```

```python
kernel = GPy.kern.Cosine(input_dim=1, active_dims=0, lengthscale=10.0) + GPy.kern.Cosine(
    input_dim=1, active_dims=1, lengthscale=2.0
)
Artist.draw_kernels(kernel, num_centers=7).plot()
```

```python
kernel = (
    GPy.kern.Matern52(input_dim=2, ARD=True, lengthscale=5)
    + GPy.kern.PeriodicExponential(input_dim=1, active_dims=0)
)
Artist.draw_kernels(kernel, num_centers=8).plot()
```
